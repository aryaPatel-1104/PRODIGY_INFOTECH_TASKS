```python
cols = df.columns
for i in range(4, len(cols)):
    df[cols[i]].fillna(df[cols[i]].mean(), inplace=True)
    
df.isna().sum().sum()
```

```python
new_df = df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
new_df.head()
```

```python
country_data = new_df[new_df['Country Name'] == "India"]
yrs = country_data.columns[1:].astype(int)
population = country_data.iloc[:, 1:].values.flatten()
plt.bar(yrs, population, width=.6, align='edge', color="red")
plt.xlabel("Years")
plt.ylabel("Population")
plt.title("Population of India over the years")
plt.show()
```

```python
country_data2 = new_df[new_df['Country Name'] == "United States"]
yrs = country_data2.columns[1:].astype(int)
population2 = country_data2.iloc[:, 1:].values.flatten()
plt.bar(yrs, population2, width=.6, align='edge', color="magenta")
plt.xlabel("Years")
plt.ylabel("Population")
plt.title("Population of USA over the years")
plt.show()
```

```python
plt.hist(population2, bins=20, edgecolor="black", color="yellow")
plt.xlabel("Population of USA")
plt.ylabel("Distribution")
plt.title("Population Distribution of USA")
plt.show()
```
